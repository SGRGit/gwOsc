```
import scipy
```

```
import matplotlib.pyplot as plt
```

```
import numpy as np
t = np.linspace(0, 60, 50)
a = np.float(np.power(1.0, -18))
fr = np.float(50.0)

x = a*np.sin(2*np.pi*fr*t) + a*np.cos(2*np.pi*fr*t)
n = np.random.randn(len(t))
xn = x + np.random.randn(len(t))
```

```
f, (ax1, ax2, ax3) = plt.subplots(1, 3)
f.set_figwidth(15)
ax1.plot(t, x, 'r' , alpha=0.75)

ax2.plot(t, n, 'b', alpha=0.75)

ax3.plot(t, xn, 'g', alpha=0.75)
```

```
from scipy.fftpack import fft, fftshift
```

```
xf = fft(x)
xnf = fft(xn)
```

```
f, (ax1, ax2) = plt.subplots(1, 2)
f.set_figwidth(20)
ax1.plot(t, xf, 'r' , alpha=0.75)
ax2.plot(t, xnf, 'g' , alpha=0.75)
```
